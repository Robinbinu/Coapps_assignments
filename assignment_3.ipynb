```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Weather Dataset
weather_data = pd.read_csv('/kaggle/input/weather-dataset/weatherHistory.csv')
```

```python
weather_data
```

```python
weather_data.shape
```

**1. describe(): Summary Statistics**

```python
weather_data.describe()
```

**2. info(): Data Types and Missing Values**

```python
weather_data.info()
```

**3. value_counts(): Counting Unique Values**

```python
weather_data['Temperature (C)'].value_counts()
```

**4. groupby(): Grouping and Aggregating**

```python
weather_data.groupby('Temperature (C)')['Humidity'].mean()
```

**5.**pivot_table(): Pivot Tables****

```python
weather_data.pivot_table('Temperature (C)', index='Humidity', columns='Apparent Temperature (C)', aggfunc='mean')
```

**6.**corr(): Correlation Matrix****

```python
weather_data.select_dtypes(include=[np.number]).corr()
```

**7. plot(): Visualizing Data**

```python
weather_data['Temperature (C)'].plot(kind='hist', title='Distribution of Temperature')
plt.show()
```

**8. rolling(): Rolling Window Calculations**

```python
weather_data['Temperature (C)'].rolling(window=10).mean()
```

**9. expanding(): Expanding Window Calculations**

```python
weather_data['Temperature (C)'].expanding().mean()
```

**10. resample(): Resampling Time Series**

```python
weather_data['Temperature (C)'].resample('M').mean()
```

**11. apply(): Applying Functions**

```python
def square(x):
    return x ** 2

weather_data['temperature_squared'] = weather_data['Temperature (C)'].apply(square)
weather_data
```

```python
weather_data['weather'] = weather_data['Summary'].map({'Partly Cloudy': 'Good Weather', 'Mostly Cloudy': 'Bad Weather'})
```
